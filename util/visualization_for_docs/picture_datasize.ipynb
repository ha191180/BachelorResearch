```python
import os
from math import floor

def get_dir_size(path='.'):
    total = 0
    with os.scandir(path) as it:
        for entry in it:
            if entry.is_file():
                total += entry.stat().st_size
            elif entry.is_dir():
                total += get_dir_size(entry.path)
    return total

denom = get_dir_size("/content/drive/MyDrive/Colab Notebooks/BachelorResearch/MER_audio_taffc_dataset_wav/2s/stft_img")/1024
print("th:", 0, "\t\t", floor(denom), "KiB\t", "100%")
data_path_tails = "0.1 0.15 0.2".split(" ")
for t in data_path_tails:
  numer = get_dir_size(data_path_head + t)/1024
  print("th:", t, "\t", floor(numer), "KiB\t", str(floor(numer*100/denom)), "%")
```

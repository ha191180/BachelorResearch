```python
from matplotlib import pyplot as plt
import pandas as pd
RESEARCH_WORK_PATH = "/content/drive/MyDrive/Colab Notebooks/BachelorResearch/"
```

```python
csv = pd.read_csv(RESEARCH_WORK_PATH + "note/data5s_0.5shift.csv")
Q12_datas = csv[(csv['label'] == "Q1") + (csv['label'] == "Q2")]
Q34_datas = csv[(csv['label'] == "Q3") + (csv['label'] == "Q4")]

for i in csv.columns:
  print(i)
  try:
    print("Q1 and Q2:")
    Q12_datas.groupby('label')[i].plot.hist(bins=50, alpha=0.7, legend=True)
    plt.show()
    print("Q3 and Q4:")
    Q34_datas.groupby('label')[i].plot.hist(bins=50, alpha=0.7, legend=True)
    plt.show()
    print("ALL")
    csv.groupby('label')[i].plot.hist(bins=50, alpha=0.7, legend=True)
    plt.show()
  except TypeError:
    print("This is not numeric array:( pass...")
    pass
```

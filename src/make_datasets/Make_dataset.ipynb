# データセット作り
ダウンロードした MER_audio_taffic_dataset_wav から、研究で利用する深層学習　感情分類用のデータセットを作成する



*   ファイル名をラベルとして扱うため、フォルダ名の感情ラベルを、元のファイル名の先頭につける。 /Q1/MT0000040632.wav　＝＞ /Q1/Q1.MT0000040632.wav

*   train valid test unknown_test に分ける
> 1.   train valid test = 8:1:1 = 720 : 90 : 90 （曲数）
  2.   30秒のデータはこれで完成


*   秒数を変えてデータの水増し (既知のデータがある）
> 1.  Q1,Q2,Q3,Q4 の各フォルダの先頭から、25曲ずつを省いておく。（未知のデータセットとして扱うため unknown_test）
  2. train valid test unknown_test に分ける 600:100:100:100
  3.   15秒の場合（30秒まるまる無い曲もある(28.5s)とか、だから最小の長さを把握してからも良いかもしれない。）
  4.   15秒に分割　/Q1/MT0000040632.wav   =>  /Q1/MT0000040632_1.wav  &  /Q1/MT0000040632_2.wav
  5.   train valid test = 8:1:1 = 1200 : 200 : 200 （曲数）

```
import os
import pathlib
import librosa
import numpy as np
from matplotlib import pyplot as plt
import cv2

def min_max(x, axis=None): # do min_max normalization ?->https://atmarkit.itmedia.co.jp/ait/articles/2110/07/news027.html
    min = 0.0
    max = 154.28702
    result = (x-min)/(max-min)
    return result

def audio_stft(filepath,n_fft = 512,hop_length = 256):
  try:
    y, sr = librosa.load(filepath)
    stft = np.abs(librosa.stft(y, n_fft=n_fft, hop_length=hop_length)) # オーバラップ率　0％
    stft = min_max(stft)
    dirs = filepath.split("/")
    output_path_str = "stft_img/" + dirs[len(dirs)-2] + '/' + dirs[len(dirs)-1] # Make end of the path ex./stft/Q1/Q1.MT0018024152_9.wav
    output_path_str = pathlib.Path(filepath).parents[1] / pathlib.Path(output_path_str) # Make head of the path ex./content/drive/My Drive/Colab Notebooks/BachelorResearch/MER_audio_taffc_dataset_wav/2s 
                                                                                        # and then, add the end of the path

  
    # print(output_path_str)
    if (stft.shape == (257, 173)):
      cv2.imwrite(str(output_path_str.parent) + '/' + output_path_str.stem + '.jpg' ,stft*255) # cv2 only support uint8(0~255) format.
    else:
      pass
    # print((stft*255).shape) -> (257, 173)
    return 
  except ValueError:
    print("An Error Accured: There is the input that has 0-size:",filepath.split("/")[len(filepath.split("/"))-1] )
    return
    
def audio_stft_unknown(filepath,n_fft = 512,hop_length = 256):
  try:
    y, sr = librosa.load(filepath)
    stft = np.abs(librosa.stft(y, n_fft=n_fft, hop_length=hop_length)) # オーバラップ率　0％
    stft = min_max(stft)
    dirs = filepath.split("/")
    output_path_str = dirs[len(dirs)-2] + '_img/' + dirs[len(dirs)-1] # Make end of the path ex./stft/Q1/Q1.MT0018024152_9.wav
    output_path_str = pathlib.Path(filepath).parents[1] / pathlib.Path(output_path_str) # Make head of the path ex./content/drive/My Drive/Colab Notebooks/BachelorResearch/MER_audio_taffc_dataset_wav/2s 
                                                                                        # and then, add the end of the path
  
    # print(output_path_str)
    if (stft.shape == (257, 173)):
      cv2.imwrite(str(output_path_str.parent) + '/' + output_path_str.stem + '.jpg' ,stft*255) # cv2 only support uint8(0~255) format.
    else:
      pass
    # print((stft*255).shape) -> (257, 173)
    return 
  except ValueError:
    print("An Error Accured: There is the input that has 0-size:",filepath.split("/")[len(filepath.split("/"))-1] )
    return
    



def path_to_audiofiles(dir_folder):
    list_of_audio = []
    for file in os.listdir(dir_folder):
        if file.endswith(".wav"):
            directory = "%s%s" % (dir_folder, file)
            list_of_audio.append(directory)
    return list_of_audio

Q1_2s = RESEARCH_WORK_PATH + "MER_audio_taffc_dataset_wav/2s/Q1/"
Q2_2s = RESEARCH_WORK_PATH + "MER_audio_taffc_dataset_wav/2s/Q2/"
Q3_2s = RESEARCH_WORK_PATH + "MER_audio_taffc_dataset_wav/2s/Q3/"
Q4_2s = RESEARCH_WORK_PATH + "MER_audio_taffc_dataset_wav/2s/Q4/"
unknown_2s = RESEARCH_WORK_PATH + "MER_audio_taffc_dataset_wav/2s/unknown/"
```

# 12/2 2sデータセット（タイムシフトver） known_2s_shift_dataset

訓練・検証・テストに分けた後、タイムシフトすることで更に水増しを行う。
28sのデータから、初めに訓練・検証・テストに使う区間を決めておく。その後、データをシフトさせて水増しする。

問題点：交差検証するのがしんどい。５回交差検証であれば、５つの場合で音楽の区間を決めて作らなければならない。

合計800曲あるので, 800✕45 = 36000曲に水増しされる

具体的なイメージ図はこの通りである。
https://app.diagrams.net/#G1dmOg1_PPDwmAInQWFCNfRjhKou4hr8nU


作成の流れ
1. １曲の音楽データを分割する。kは 交差検証k回を表す. \\
  k=1 のとき　train (0~20s) val (20s ~ 24s) test (24s ~ 28s) を用いる. \\
  Train : i = (0,36)  Val : i = (40,44)  Test : i = (48,52) \\
2.  リスト項目

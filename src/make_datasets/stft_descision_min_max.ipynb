```python
from librosa.core.convert import midi_to_note
import os
import pathlib
import librosa
import numpy as np
from matplotlib import pyplot as plt
import cv2


def audio_max_magnitude(filepath,n_fft = 512,hop_length = 256): # 154.28702
  try:
    y, sr = librosa.load(filepath)
    stft = np.abs(librosa.stft(y, n_fft=n_fft, hop_length=hop_length)) # オーバラップ率　0％
    return stft.max()
  except Exception:
    return 0

def audio_min_magnitude(filepath,n_fft = 512,hop_length = 256): # -inf
  try:
    y, sr = librosa.load(filepath)
    stft = np.abs(librosa.stft(y, n_fft=n_fft, hop_length=hop_length)) # オーバラップ率　0％
    return stft.min()
  except Exception:
    return 0

    
def path_to_audiofiles(dir_folder):
    list_of_audio = []
    for file in os.listdir(dir_folder):
        if file.endswith(".wav"):
            directory = "%s%s" % (dir_folder, file)
            list_of_audio.append(directory)
    return list_of_audio

Q1_5s = "/content/data/Q1/"
Q2_5s = "/content/data/Q2/"
Q3_5s = "/content/data/Q3/"
Q4_5s = "/content/data/Q4/"

# Search the maxium value of arrays
Q1_max = -1
for audio in path_to_audiofiles(Q1_5s):
  tmp = audio_max_magnitude(audio)
  if (tmp > Q1_max):
    Q1_max = tmp
Q2_max = -1
for audio in path_to_audiofiles(Q2_5s):
  tmp = audio_max_magnitude(audio)
  if (tmp > Q2_max):
    Q2_max = tmp
Q3_max = -1
for audio in path_to_audiofiles(Q3_5s):
  tmp = audio_max_magnitude(audio)
  if (tmp > Q3_max):
    Q3_max = tmp
Q4_max = -1
for audio in path_to_audiofiles(Q4_5s):
  tmp = audio_max_magnitude(audio)
  if (tmp > Q4_max):
    Q4_max = tmp

max_of_all = max(Q1_max,Q2_max,Q3_max,Q4_max)
print(max_of_all)

Q1_min = 1e9
for audio in path_to_audiofiles(Q1_5s):
  tmp = audio_min_magnitude(audio)
  if (tmp < Q1_min):
    Q1_min = tmp
Q2_min = 1e9
for audio in path_to_audiofiles(Q2_5s):
  tmp = audio_min_magnitude(audio)
  if (tmp < Q2_min):
    Q2_min = tmp
Q3_min = 1e9
for audio in path_to_audiofiles(Q3_5s):
  tmp = audio_min_magnitude(audio)
  if (tmp < Q3_min):
    Q3_min = tmp
Q4_min = 1e9
for audio in path_to_audiofiles(Q4_5s):
  tmp = audio_min_magnitude(audio)
  if (tmp < Q4_min):
    Q4_min = tmp

min_of_all = min(Q1_min,Q2_min,Q3_min,Q4_min)
print(min_of_all)
```

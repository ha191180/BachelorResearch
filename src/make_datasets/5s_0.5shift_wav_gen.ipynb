# Prepare Directories

```python
import os

DATASET_PATH = "/content/drive/MyDrive/Colab Notebooks/BachelorResearch/MER_audio_taffc_dataset_wav/renamed/"
DST_PATH = "/content/drive/MyDrive/Colab Notebooks/BachelorResearch/MER_audio_taffc_dataset_wav/5s_0.5shift/wav/"
DST_PATH_UNKNOWN = "/content/drive/MyDrive/Colab Notebooks/BachelorResearch/MER_audio_taffc_dataset_wav/5s_0.5shift/unknown/wav/"

if not os.path.exists(DST_PATH):
  os.makedirs(DST_PATH)

Qs = "Q1 Q2 Q3 Q4".split(" ")
for q in Qs:
  if not os.path.exists(DST_PATH + q):
    os.makedirs(DST_PATH + q)

if not os.path.exists(DST_PATH_UNKNOWN):
  os.makedirs(DST_PATH_UNKNOWN)
```

# Slicing waves

```python
import librosa
import pathlib
import soundfile as sf
from tqdm import tqdm


def slice_wav(audio_path, dst_path_root):
  y, sr = librosa.load(audio_path)

  q = audio_path.stem.split(".")[0]
  save_dir = pathlib.Path(dst_path_root + q)

  for i in range(46):
    start = sr//2*i
    stop  = 5*sr + sr//2*i
    s = y[start:stop]
    filename = audio_path.stem + "_" + str(i) + audio_path.suffix
    path = save_dir / filename
    sf.write(str(path), s, sr)
  return
    

def slice_wav_unknown(audio_path, dst_path_root):
  y, sr = librosa.load(audio_path)

  save_dir = pathlib.Path(dst_path_root)

  for i in range(46):
    start = sr//2*i
    stop  = 5*sr + sr//2*i
    s = y[start:stop]
    filename = audio_path.stem + "_" + str(i) + audio_path.suffix
    path = save_dir / filename
    sf.write(str(path), s, sr)
  return
    


for q in pathlib.Path(DATASET_PATH).glob("*"):
  if "Q" in str(q):
    for audio_path in tqdm(q.glob("*.wav")):
      slice_wav(audio_path, DST_PATH)
      

      
  else:
    for audio_path in tqdm(q.glob("*.wav")):
      slice_wav_unknown(audio_path, DST_PATH_UNKNOWN)
```

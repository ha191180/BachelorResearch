```python
import os
import pathlib
import librosa
import numpy as np

def min_max(x, axis=None): # do min_max normalization ?->https://atmarkit.itmedia.co.jp/ait/articles/2110/07/news027.html
    min = 0
    max = 154.28702 
    # max = 137.74599
    x = np.abs(x)
    result = (x-min)/(max-min)
    return result

def to_db(x, dBref=2e-5):
    y = 20 * np.log10(x / dBref)
    return y

def audio_stft_crop(filepath,crop_position=224,n_fft = 512,hop_length = 256):
  try:
    add_div = 4
    y, sr = librosa.load(filepath)
    stft = np.abs(librosa.stft(y, n_fft=n_fft, hop_length=hop_length)) # オーバラップ率　0％
    stft = min_max(stft)
    x =  stft[:crop_position,:]

    ret = np.empty((0, x.shape[1], add_div),int)
    for i in x:
      tmp = np.empty((0, add_div), int)

      for j in i:
        tmp2 = []

        for k in range(add_div):
          tmp2.append(j)

        tmp2 = np.array(tmp2)
        tmp2 = tmp2.reshape((1,add_div))

        tmp = np.append(tmp, tmp2, axis=0)

      tmp = tmp.reshape((1, x.shape[1], add_div))
      ret = np.append(ret,tmp,axis=0)
    return ret
  except Exception:
    return
    
def path_to_audiofiles(dir_folder):
    list_of_audio = []
    for file in os.listdir(dir_folder):
        if file.endswith(".wav"):
            directory = "%s%s" % (dir_folder, file)
            list_of_audio.append(directory)
    return list_of_audio

Q1_2s = RESEARCH_WORK_PATH + "MER_audio_taffc_dataset_wav/2s/Q1/"
Q2_2s = RESEARCH_WORK_PATH + "MER_audio_taffc_dataset_wav/2s/Q2/"
Q3_2s = RESEARCH_WORK_PATH + "MER_audio_taffc_dataset_wav/2s/Q3/"
Q4_2s = RESEARCH_WORK_PATH + "MER_audio_taffc_dataset_wav/2s/Q4/"
```

```python
l = path_to_audiofiles(Q1_2s)
ll = len(l)

a1 = audio_stft_crop(l[0])
x = np.empty((0,a1.shape[0], a1.shape[1],a1.shape[2]), int)
y = []
for i,audio in enumerate(l):
  xtmp = audio_stft_crop(audio)
  xtmp = xtmp.reshape((1,xtmp.shape[0], xtmp.shape[1], xtmp.shape[2]))
  x = np.append(x, xtmp, axis=0)
  print("\r", (i+1)*100//ll, "%", end='')
  y.append(1)
```

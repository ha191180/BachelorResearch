```python
RESEARCH_WORK_DIR = "/content/drive/MyDrive/Colab Notebooks/BachelorResearch/"
Q1_dir = RESEARCH_WORK_DIR + "MER_audio_taffc_dataset_wav/2s/Q1/"
Q2_dir = RESEARCH_WORK_DIR + "MER_audio_taffc_dataset_wav/2s/Q2/"
Q3_dir = RESEARCH_WORK_DIR + "MER_audio_taffc_dataset_wav/2s/Q3/"
Q4_dir = RESEARCH_WORK_DIR + "MER_audio_taffc_dataset_wav/2s/Q4/"

src_dirs = [Q1_dir, Q2_dir, Q3_dir, Q4_dir]

Q1_save_dir = "/content/drive/MyDrive/Colab Notebooks/BachelorResearch/MER_audio_taffc_dataset_wav/2s/log-melspectro_img/Q1/"
Q2_save_dir = "/content/drive/MyDrive/Colab Notebooks/BachelorResearch/MER_audio_taffc_dataset_wav/2s/log-melspectro_img/Q2/"
Q3_save_dir = "/content/drive/MyDrive/Colab Notebooks/BachelorResearch/MER_audio_taffc_dataset_wav/2s/log-melspectro_img/Q3/"
Q4_save_dir = "/content/drive/MyDrive/Colab Notebooks/BachelorResearch/MER_audio_taffc_dataset_wav/2s/log-melspectro_img/Q4/"

dst_dirs = [Q1_save_dir, Q2_save_dir, Q3_save_dir, Q4_save_dir]
```

```python
for d in dst_dirs:
  if not os.path.exists(d):
    os.makedirs(d)
```

```python
def path_to_audiofiles(dir_folder):
    list_of_audio = []
    for file in os.listdir(dir_folder):
        if file.endswith(".wav"):
            directory = "%s%s" % (dir_folder, file)
            list_of_audio.append(directory)
    return list_of_audio
```

```python
def min_max(x, axis=None): # do min_max normalization ?->https://atmarkit.itmedia.co.jp/ait/articles/2110/07/news027.html
    min = -80.0
    max = 0
    result = (x-min)/(max-min)
    return result
```

```python
def melspec(audio):
  y, sr = librosa.load(audio)
  d = librosa.feature.melspectrogram(y,sr,n_fft=512,hop_length=256, power=2)
  d = librosa.amplitude_to_db(d, ref=np.max)
  d = min_max(d)
  return d
```

```python
for _ in range(4):
  src = src_dirs[_]
  dst = dst_dirs[_]

  for audio in tqdm(path_to_audiofiles(src)):
    y = melspec(audio)
    cv2.imwrite(dst + pathlib.Path(audio).stem + '.jpg' ,y*255) # cv2 only support uint8(0~255) format.
```

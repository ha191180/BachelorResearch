```python
import pickle
from matplotlib import pyplot as plt
from glob import glob
import pathlib

leg = []
for f in glob('/content/drive/MyDrive/BachelorResearch/src/train_my_model_stft_1d_features_3/datas/*'):
  with open(f, 'rb') as p:
      datas = pickle.load(p)
      acc_hist = datas['acc']
      loss_hist = datas['loss']
      plt.plot([a.cpu() for a in acc_hist[0]])
      # plt.plot([a.cpu() for a in acc_hist[1]])
      leg.append(pathlib.Path(f).stem)
plt.title('train accuracy')
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.legend(leg, loc='lower right')
plt.show()

leg = []
for f in glob('/content/drive/MyDrive/BachelorResearch/src/train_my_model_stft_1d_features_3/datas/*'):
  with open(f, 'rb') as p:
      datas = pickle.load(p)
      acc_hist = datas['acc']
      loss_hist = datas['loss']
      plt.plot([a.cpu() for a in acc_hist[0]])
      # plt.plot([a.cpu() for a in acc_hist[1]])
      leg.append(pathlib.Path(f).stem)
plt.title('test accuracy')
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.legend(leg, loc='lower right')
plt.show()

leg = []
for f in glob('/content/drive/MyDrive/BachelorResearch/src/train_my_model_stft_1d_features_3/datas/*'):
  with open(f, 'rb') as p:
      datas = pickle.load(p)
      acc_hist = datas['acc']
      loss_hist = datas['loss']
      plt.plot(loss_hist[0])
      leg.append(pathlib.Path(f).stem)
plt.title('train loss')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend(leg, loc='lower right')
plt.show()


leg = []
for f in glob('/content/drive/MyDrive/BachelorResearch/src/train_my_model_stft_1d_features_3/datas/*'):
  with open(f, 'rb') as p:
      datas = pickle.load(p)
      acc_hist = datas['acc']
      loss_hist = datas['loss']
      plt.plot(loss_hist[1])
      leg.append(pathlib.Path(f).stem)
plt.title('test loss')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend(leg, loc='lower right')
plt.show()
```

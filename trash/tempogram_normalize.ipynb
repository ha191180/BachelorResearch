```python
import librosa
import numpy as np
from tqdm import tqdm
from matplotlib import pyplot as plt


for q in "Q1 Q2 Q3 Q4".split(" "):
  d = data_dir + q + "/"
  q = int(q[1])-1
  songlist = os.listdir(d)
  a = songlist[100]
  audio_path = os.path.join(d, a)
  audio, sr = librosa.load(audio_path)
  t = librosa.feature.tempogram(audio, sr=sr)
  print(t.shape)
  plt.imshow(t, cmap='gray')
  print(t.max())
  plt.show()
```

```python
import librosa
import numpy as np
from tqdm import tqdm

tempograms = [[], [], [], []]


for q in "Q1 Q2 Q3 Q4".split(" "):
  d = data_dir + q + "/"
  max = -1e9
  min = 1e9
  for a in tqdm(os.listdir(d)):
    audio_path = os.path.join(d, a)
    audio, sr = librosa.load(audio_path)
    t = librosa.feature.tempogram(audio, sr=sr)
    if max < t.max():
      max = t.max()
    if min > t.min():
      min = t.min()
  print(min)
  print(max)

```

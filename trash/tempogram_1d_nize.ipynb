```python
import librosa
import numpy as np
from tqdm import tqdm

tempograms = [[], [], [], []]


for q in "Q1 Q2 Q3 Q4".split(" "):
  ma = -1e9
  mi = 1e9
  d = data_dir + q + "/"
  for a in tqdm(os.listdir(d)):
    audio_path = os.path.join(d, a)
    audio, sr = librosa.load(audio_path)
    t = librosa.feature.tempogram(audio, sr=sr)
    f1d = []
    for i in range(len(t[0])):
      tmp = 0
      for j in range(len(t)):
        tmp += t[j,i] * j
      tmp = tmp/len(t)
      f1d.append(tmp)
    if ma < max(f1d):
      ma = max(f1d)
    if mi > min(f1d):
      mi = min(f1d)
  print(q)
  print(ma)
  print(mi)
```
